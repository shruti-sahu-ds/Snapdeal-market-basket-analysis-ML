```python
import pandas as pd 
def load_data() : 
    df= pd.read_csv("SnapDeal.csv")
    return df
```

```python
def remove_duplicates(df) :
    df = df.drop_duplicates()
    return df
```

```python
def clean_columns(df):
    df.columns = df.columns.str.strip()   ## remove extra spaces 
    df.columns = df.columns.str.replace(" ","_")  ## remove space with _
    return df
```

```python
def handle_missing_values(df):
    df = df.fillna("Unknown")   #fill null values with 'Unknown'
    return df   
    
```

```python
def standardize_data(df):
    # example for gender 
    if 'Gender' in df.columns:
        df['Gender'] = df['Gender'].str.lower().str.strip()
        #example for recommendation 
    if 'Recommendation' in df.columns: 
        df['Recommendation'] = df['Recommendation'].str.lower().str.strip()

    return(df)
        
```

```python
def convert_rating(df):
    rating_cols = ['Customer_Reviews_Importance',
                  'Rating_Accuracy',
                  'Delivery_Speed_Importance']
    for col in rating_cols : 
        if col in df.columns:
            df[col] = pd.to_numeric(df[col], errors = 'coerce')
    return df
```

```python
def clean_data(df):
    df = remove_duplicates(df)
    df = clean_columns(df)
    df = handle_missing_values(df)
    df = standardize_data(df)
    df = convert_rating(df)
    return df
                        
```

```python
df = load_data()
df = clean_data(df)
df.head()
```
