```python
import pandas as pd
def load_data():
    return pd.read_csv("SnapDeal.csv")
    
def clean_data(df):
    return df
    
df = load_data()
df = clean_data(df)
df.head()
```

### Customer Demographics

## 1.Gender Distribution

```python
print (df['Gender'] .value_counts())

import seaborn as sns 
import matplotlib.pyplot as plt

sns.countplot(x='Gender',data=df)
plt.title("Gender Distribution")
plt.show()
```

## 2.Age_Distribution

```python
print(df['age'].describe())

sns.histplot(df['age'] ,kde = True)
plt.title ("Age Distribution")
plt.show()
```

### Purchase Frequency & Product Categories 

## 1.Purchase Frequency 

```python
print(df['Purchase_Frequency'].value_counts())

sns.countplot(x='Purchase_Frequency',data=df)
plt.xticks(rotation=45)
plt.title("Purchase_Frequency")
plt.show()
```

## 2. ProductCategory

```python
top10 = df ['Purchase_Categories'] .value_counts().nlargest(10).index
print(df['Purchase_Categories'] . value_counts())

sns.countplot(y='Purchase_Categories',data=df,order=top10)
    
plt.title("Purchase_Categories")
plt.show()
```

## 3. Browsing_Methods

```python

print(df['Product_Search_Method'] .value_counts())

sns.countplot(y = 'Product_Search_Method',data = df, color = 'teal' )
plt.title('Browsing_Methods')
plt.show()
```

## 4. Cart Abdonment factors 

```python
print(df['Cart_Abandonment_Factors'].value_counts())

sns.countplot(y = 'Cart_Abandonment_Factors' , data = df,color = 'forestgreen')

plt.title('Cart_Abandonment_Factors')
plt.show()
```

## 5.Mean & Median 

```python
mapping = {
    "Yes" : 1,
    "No"  : 2,
    "Sometimes" : 3
    }

df['Recommendation_Helpfulness'] = df['Recommendation_Helpfulness'] .map(mapping)
```

```python
print("Mean_Values:")
print(df[['Recommendation_Helpfulness',
       'Rating_Accuracy ', 
    'Shopping_Satisfaction']].mean().round(0))

print("\nMedian_Values: ")
print(df[['Recommendation_Helpfulness',
       'Rating_Accuracy ',
          'Shopping_Satisfaction']].median().round(0))
```
