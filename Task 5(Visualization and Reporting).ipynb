### Task 5-Visualization and Reporting
### Task 5.1: Purchase category distribution

We visualize the most frequently selected product categories
to understand customer buying interests.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# load cleaned dataset
df = pd.read_csv("SnapDeal_Cleaned.csv")

df.head()
```

```python
# split multi-category selections
categories = df['Purchase_Categories'].str.split(';').explode().str.strip()

categories.value_counts().head(10).plot(kind='bar')
plt.title("Top Product Categories")
plt.xlabel("Category")
plt.ylabel("Number of Selections")
plt.xticks(rotation=45)
plt.show()
```

### Task 5.2: Browsing frequency distribution

We analyze how frequently customers browse the platform.

```python
df['Browsing_Frequency'].value_counts().plot(kind='bar')
plt.title("Browsing Frequency Distribution")
plt.xlabel("Frequency")
plt.ylabel("Customers")
plt.xticks(rotation=45)
plt.show()
```

### Task 5.3: Customer satisfaction levels

We visualize overall satisfaction ratings across customers.

```python
df['Shopping_Satisfaction'].value_counts().sort_index().plot(kind='bar')
plt.title("Shopping Satisfaction Levels")
plt.xlabel("Satisfaction Rating")
plt.ylabel("Customers")
plt.show()
```

### Task 5.4: Recommendation usefulness vs satisfaction

We visualize how recommendation helpfulness relates to customer satisfaction.

```python
df.groupby('Recommendation_Helpfulness')['Shopping_Satisfaction'].mean().plot(kind='bar')
plt.title("Recommendation Helpfulness vs Satisfaction")
plt.xlabel("Helpfulness")
plt.ylabel("Average Satisfaction")
plt.show()
```
