### Task 4: Recommendation and Review Insights
#### Task 4.1: Relationship between recommendation helpfulness and shopping satisfaction

We analyze how customers' perception of recommendation helpfulness
relates to their overall shopping satisfaction.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# load cleaned dataset
df = pd.read_csv("SnapDeal_Cleaned.csv")

df.head()
```

```python
df.groupby('Recommendation_Helpfulness')['Shopping_Satisfaction'].mean()
```

```python
df.groupby('Recommendation_Helpfulness')['Shopping_Satisfaction'].mean().plot(kind='bar')
plt.title("Recommendation Helpfulness vs Shopping Satisfaction")
plt.xlabel("Recommendation Helpfulness")
plt.ylabel("Average Satisfaction")
plt.show()
```

### Task 4.2: Impact of review reliability and helpfulness on ratings

We examine whether trust in reviews influences rating accuracy and satisfaction.

```python
df.groupby('Review_Reliability')['Rating_Accuracy'].mean()
```

```python
df.groupby('Review_Reliability')['Shopping_Satisfaction'].mean()
```

### Task 4.3: Engagement with personalized recommendations

We analyze how frequently customers engage with personalized recommendations.

```python
df['Personalized_Recommendation_Frequency'].value_counts()
```

```python
df['Personalized_Recommendation_Frequency'].value_counts().plot(kind='bar')
plt.title("Personalized Recommendation Engagement")
plt.xlabel("Frequency")
plt.ylabel("Customers")
plt.xticks(rotation=45)
plt.show()
```

### Task 4.4: Actionable insights

Based on the above analysis, we derive insights to improve recommendation quality
and customer engagement.

###  Actionable insights

• Recommendation helpfulness has only a small effect on satisfaction.
Customers who said recommendations are "Sometimes" helpful have the highest satisfaction (~3.08),
while "Yes" and "No" users are slightly lower (~3.02 and ~2.89).  
This suggests recommendations exist but are not strongly influencing decisions.

• Review reliability does not significantly change customer experience.
Rating accuracy and satisfaction stay close to 3 across all trust levels,
meaning customers are not heavily depending on reviews while purchasing.

• Personalized recommendations are underused.
Most users either do not use them (285) or only sometimes use them (272),
while fewer actively rely on them (243).  
This indicates low engagement with the recommendation system.

• Business implication:
Improve relevance of recommendations and highlight trustworthy reviews
(verified buyers, top helpful reviews) to increase engagement and satisfaction.
